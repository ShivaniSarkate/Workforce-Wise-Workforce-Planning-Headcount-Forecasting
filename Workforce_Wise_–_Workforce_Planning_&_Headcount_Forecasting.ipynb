**Project Title : WorkforceWise - Workforce Planning & Headcount Forecasting**

**Objective** :Forecast workforce needs based on business growth and attrition trends to support strategic hiring and budgeting.

**Business Context** : HR and operations teams need to anticipate staffing needs to avoid under-resourcing or over-hiring. Forecasting helps align headcount with business goals.

**Workflow Overview** :
- Simulate historical headcount and attrition  
- Model business growth scenarios  
- Forecast future headcount needs  
- Visualize hiring gaps  
- Recommend workforce strategy

```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Simulate headcount data
np.random.seed(42)
months = pd.date_range(start='2023-01-01', periods=24, freq='M')
headcount = np.linspace(100, 160, 24) + np.random.normal(0, 5, 24)
attrition_rate = np.random.uniform(0.02, 0.05, 24)
growth_rate = 0.04  # monthly business growth

df = pd.DataFrame({'Month': months, 'Headcount': headcount, 'AttritionRate': attrition_rate})
df['Attrition'] = df['Headcount'] * df['AttritionRate']
df['ProjectedNeed'] = df['Headcount'] * (1 + growth_rate)

# Step 3: Visualize
plt.figure(figsize=(10,6))
plt.plot(df['Month'], df['Headcount'], label='Current Headcount')
plt.plot(df['Month'], df['ProjectedNeed'], label='Projected Need', linestyle='--')
plt.fill_between(df['Month'], df['Headcount'], df['ProjectedNeed'], color='orange', alpha=0.3, label='Hiring Gap')
plt.title('Workforce Planning Forecast')
plt.xlabel('Month')
plt.ylabel('Employees')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Executive Summary** :
- **Projected Gap**: ~25 employees by Month 24  
- **Attrition Impact**: ~35% monthly loss  
- **Recommendation**:  
  - Hire proactively to meet projected demand  
  - Monitor attrition trends quarterly  
  - Align workforce planning with revenue forecasts  
- **Impact**: Forecasting headcount needs improves operational readiness and budgeting accuracy
